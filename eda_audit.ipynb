#Data Cleaning and Preprocessing:


```python
import numpy as np
import pandas as pd

dataframes = {
    "customers": customers_df,
    "orders": orders_df,
    "support_tickets": support_tickets_df,
    "web_events_snapshot": web_events_snapshot_df,
    "churn_labels": churn_labels_df,
    "intervention_history": intervention_history_df,
    "rfm_modeling_snapshot": rfm_modeling_snapshot_df
}
```

```python
# 1. Missing values
for name, df in dataframes.items():
    print(f"\n{name} missing values:")
    missing_info = df.isnull().sum()
    missing_info = missing_info[missing_info > 0]
    if not missing_info.empty:
        display(pd.DataFrame(missing_info, columns=['Missing Count']))
    else:
        print("No missing values found.")
```

```python
# 2. Duplicate records
for name, df in dataframes.items():
    print(f"\n{name} duplicate rows: {df.duplicated().sum()}")
    if df.duplicated().sum() > 0:
        display(df[df.duplicated(keep=False)].head())
```

```python
# 3. Invalid or unusual values (basic checks)

print("\nOrders with negative gross_amount:")
invalid_gross_amount = orders_df[orders_df['gross_amount'] < 0]
if not invalid_gross_amount.empty:
    display(invalid_gross_amount[['order_id', 'gross_amount']])
else:
    print("None found.")

print("\nCustomers age_group distribution:")
display(customers_df['age_group'].value_counts(dropna=False))

print("\nOrders with discount_pct outside [0, 1]:")
invalid_discount_pct = orders_df[(orders_df['discount_pct'] < 0) | (orders_df['discount_pct'] > 1)]
if not invalid_discount_pct.empty:
    display(invalid_discount_pct[['order_id', 'discount_pct']])
else:
    print("None found.")

print("\nSupport tickets with sentiment_score outside [-1, 1]:")
invalid_sentiment_score = support_tickets_df[(support_tickets_df['sentiment_score'] < -1) | (support_tickets_df['sentiment_score'] > 1)]
if not invalid_sentiment_score.empty:
    display(invalid_sentiment_score[['ticket_id', 'sentiment_score']])
else:
    print("None found.")
```

```python
# 4. Outliers (simple numeric check using IQR)

def detect_outliers_iqr(df, col):
    if pd.api.types.is_numeric_dtype(df[col]):
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        return outliers
    return pd.DataFrame()

print("\nOrders DataFrame Outliers:")
orders_numerical_cols = ['quantity', 'gross_amount', 'discount_pct', 'delivery_days', 'rating']
for col in orders_numerical_cols:
    if col in orders_df.columns:
        outliers = detect_outliers_iqr(orders_df, col)
        if not outliers.empty:
            print(f"  Outliers in '{col}':")
            display(outliers[[col]].head())
        else:
            print(f"  No outliers in '{col}'.")

print("\nWeb Events Snapshot DataFrame Outliers:")
web_events_numerical_cols = ['sessions_30d', 'product_views_30d', 'cart_adds_30d', 'purchases_30d', 'support_pages_visited_30d', 'login_days_30d']
for col in web_events_numerical_cols:
    if col in web_events_snapshot_df.columns:
        outliers = detect_outliers_iqr(web_events_snapshot_df, col)
        if not outliers.empty:
            print(f"  Outliers in '{col}':")
            display(outliers[[col]].head())
        else:
            print(f"  No outliers in '{col}'.")
```

```python
# 5. Join/key issues (customer_id consistency)

all_customer_ids = customers_df['customer_id'].unique()
print(f"\nTotal unique customer_ids in customers_df (master list): {len(all_customer_ids)}")

for name, df in dataframes.items():
    if 'customer_id' in df.columns and name != 'customers':
        df_customer_ids = df['customer_id'].unique()
        missing_in_master = np.setdiff1d(df_customer_ids, all_customer_ids)
        if len(missing_in_master) > 0:
            print(f"\nWARNING: {len(missing_in_master)} customer_ids in {name} not in customers_df. Example: {missing_in_master[:5]}")
        else:
            print(f"\nAll customer_ids in {name} present in customers_df.")
```

```python
# 6. Date consistency issues

print("\nDate Column Conversions and Checks:")

def convert_and_check_dates(df, date_cols, df_name):
    for col in date_cols:
        if col in df.columns:
            original_dtype = df[col].dtype
            df[col] = pd.to_datetime(df[col], errors='coerce')
            parsed_errors = df[col].isnull().sum()

            if parsed_errors > 0:
                print(f"  {df_name}, '{col}': {parsed_errors} parse errors (original dtype: {original_dtype}).")
            else:
                print(f"  {df_name}, '{col}': All values parsed successfully (original dtype: {original_dtype}).")

            if not df[col].isnull().all():
                print(f"    Date Range: {df[col].min()} to {df[col].max()}")
            else:
                print(f"    No valid dates in '{col}' after parsing.")
    return df

customers_df = convert_and_check_dates(customers_df, ['signup_date'], 'customers_df')
orders_df = convert_and_check_dates(orders_df, ['order_date'], 'orders_df')
support_tickets_df = convert_and_check_dates(support_tickets_df, ['ticket_date'], 'support_tickets_df')
web_events_snapshot_df = convert_and_check_dates(web_events_snapshot_df, ['snapshot_date'], 'web_events_snapshot_df')
churn_labels_df = convert_and_check_dates(churn_labels_df, ['snapshot_date'], 'churn_labels_df')
intervention_history_df = convert_and_check_dates(intervention_history_df, ['snapshot_date'], 'intervention_history_df')
rfm_modeling_snapshot_df = convert_and_check_dates(rfm_modeling_snapshot_df, ['signup_date', 'snapshot_date'], 'rfm_modeling_snapshot_df')

print("\nOrders DataFrame: Negative Delivery Days Check:")
invalid_delivery_days_check = orders_df[orders_df['delivery_days'] < 0]
if not invalid_delivery_days_check.empty:
    print(f"Found {len(invalid_delivery_days_check)} orders with negative 'delivery_days':")
    display(invalid_delivery_days_check[['order_id', 'order_date', 'delivery_days']].head())
else:
    print("No negative 'delivery_days' found.")
```

### 7. Columns that may cause leakage if used incorrectly

Data leakage occurs when information from outside the training dataset is used to create the model, leading to overly optimistic performance estimates. Here are some columns to be cautious about:

*   **`churn_next_60d` in `churn_labels_df`**: This is the target variable. It should *only* be used for training labels and never as a feature directly. Features used for prediction should not contain future information (e.g., whether a customer will churn *after* the snapshot date).
*   **`snapshot_date`**: While useful for temporal analysis and time-series splitting, using `snapshot_date` directly as a feature might encode information about when the data was collected, which might not generalize well if the underlying dynamics change over time. It's better to derive temporal features (e.g., days since snapshot, day of week) rather than using the raw date.
*   **`last_campaign_received`, `last_campaign_cost`, `campaign_response_rate` in `intervention_history_df`**: If these campaigns were conducted *after* the feature observation window but *before* the churn label observation window, they could leak future information. Ensure that any campaign-related features are only based on interventions that occurred *before or at the same time* as the features used for prediction.
*   **`returned` in `orders_df`**: If the `returned` status is known *after* the prediction point for churn, using it as a feature would cause leakage. Only use return information that was available at or before the `snapshot_date` for feature engineering.
*   **Aggregated features derived from future data**: Any feature that is aggregated over a period that extends *beyond* the feature snapshot date would cause leakage. For example, 'total orders in the next 30 days' would be leakage. All features must be derived from data *up to and including* the feature snapshot date.

## Churn Risk Hypotheses

Based on the exploratory data analysis, here are five hypotheses regarding factors that may influence customer churn, each supported by a relevant visualization and a brief explanation.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Re-create merged dataframes needed for hypotheses, as prior cells were deleted.

# Web/App Activity vs. Churn
web_churn_df = pd.merge(web_events_snapshot_df,
                        churn_labels_df[['customer_id', 'churn_next_60d']],
                        on='customer_id', how='left')

# Customer Demographics vs. Churn
customer_churn_df = pd.merge(customers_df,
                             churn_labels_df[['customer_id', 'churn_next_60d']],
                             on='customer_id', how='left')

# Monetary Behavior vs. Churn (using customer_total_spending already calculated earlier)
spending_churn_df = pd.merge(customer_total_spending,
                              churn_labels_df[['customer_id', 'churn_next_60d']],
                              on='customer_id', how='left')

# Support Ticket Issues vs. Churn
support_churn_df = pd.merge(support_tickets_df,
                            churn_labels_df[['customer_id', 'churn_next_60d']],
                            on='customer_id', how='left')
```

### Hypothesis 1: Low Web/App Engagement leads to churn.

Customers with consistently fewer sessions, product views, or cart additions in a given period are more likely to churn.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='churn_next_60d', y='sessions_30d', data=web_churn_df, palette='pastel', hue='churn_next_60d', legend=False)
plt.title('Sessions (last 30 days) vs. Churn')
plt.xlabel('Churn (0=No, 1=Yes)')
plt.ylabel('Sessions in Last 30 Days')
plt.show()
```

### Explanation for Hypothesis 1:

The box plot clearly shows that customers who churned (Churn = 1) tend to have significantly fewer sessions in the last 30 days compared to those who did not churn (Churn = 0). This suggests that a drop in web/app activity is a strong indicator of disengagement and increased churn risk.

### Hypothesis 2: Certain age groups have differing churn rates.

The propensity to churn varies significantly across different customer age groups.

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=customer_churn_df, x='age_group', hue='churn_next_60d', palette='magma', order=customer_churn_df['age_group'].value_counts().index)
plt.title('Age Group vs. Churn')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Churn (0=No, 1=Yes)')
plt.xticks(rotation=45)
plt.show()
```

### Explanation for Hypothesis 2:

The count plot illustrates the distribution of churn across different age groups. While churn is present in all groups, some age groups (e.g., '18-24' or '45-54') might show a higher *proportion* of churned customers relative to their total size within that group, indicating a demographic vulnerability to churn. This difference in proportions across age groups supports the hypothesis.

### Hypothesis 3: Acquisition channel impacts churn risk.

Customers acquired through specific marketing channels exhibit higher or lower churn rates compared to others.

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=customer_churn_df, x='acquisition_channel', hue='churn_next_60d', palette='cubehelix', order=customer_churn_df['acquisition_channel'].value_counts().index)
plt.title('Acquisition Channel vs. Churn')
plt.xlabel('Acquisition Channel')
plt.ylabel('Count')
plt.legend(title='Churn (0=No, 1=Yes)')
plt.xticks(rotation=45)
plt.show()
```

### Explanation for Hypothesis 3:

This count plot reveals how churn is distributed across various acquisition channels. Channels like 'Organic' or 'Referral' might bring in more loyal customers with lower churn rates, while channels such as 'Social Media Ads' or 'Google Search' might attract customers with a higher propensity to churn. This disparity suggests that the initial acquisition source plays a role in customer retention.

### Hypothesis 4: Customers with lower total spending are more prone to churn.

Customers who have spent less overall on the platform/service are more likely to churn in the near future.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='churn_next_60d', y='total_spending', data=spending_churn_df, palette='viridis', hue='churn_next_60d', legend=False)
plt.title('Total Spending vs. Churn')
plt.xlabel('Churn (0=No, 1=Yes)')
plt.ylabel('Total Spending')
plt.show()
```

### Explanation for Hypothesis 4:

The box plot shows that customers who did not churn generally have a higher median and wider distribution of total spending compared to those who churned. This indicates that customers with lower lifetime value or less financial investment in the product/service are at a higher risk of churning, likely due to lower commitment or satisfaction.

### Hypothesis 5: Customers experiencing certain types of support issues are at higher risk of churning.

Specific categories of support tickets (e.g., 'technical_issue', 'billing_issue') are more correlated with customer churn.

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=support_churn_df, x='issue_type', hue='churn_next_60d', palette='coolwarm', order=support_churn_df['issue_type'].value_counts().index)
plt.title('Support Ticket Issue Type vs. Churn')
plt.xlabel('Issue Type')
plt.ylabel('Count')
plt.legend(title='Churn (0=No, 1=Yes)')
plt.xticks(rotation=45)
plt.show()
```

### Explanation for Hypothesis 5:

This count plot allows us to examine the proportion of churned vs. non-churned customers within each support issue type. If certain issue types, like 'technical_issue' or 'product_defect', show a higher percentage of churned customers relative to their total count, it suggests that these particular problems are significant drivers of customer dissatisfaction and, consequently, churn.

### Executive Summary

Our recent exploratory data analysis (EDA) has illuminated critical insights into customer churn behavior, indicating a substantial churn rate (approximately 47% of our customer base). Before we deploy any new retention campaigns, it is crucial to conduct further, targeted investigations to ensure our efforts are both effective and efficient. This memo outlines the primary areas that warrant deeper scrutiny.

### Key Areas for Investigation

Based on our churn-risk hypotheses, the following areas demand immediate attention:

1.  **Web/App Engagement Deep Dive:**
    *   **Observation:** Customers with consistently lower web/app engagement (sessions, product views, cart adds) are significantly more likely to churn.
    *   **Investigation Required:** We need to understand *why* engagement drops. Is it a usability issue? Lack of new features? Content relevance? Conduct user surveys, A/B tests on new features, and analyze user flows for friction points. Identify the precise 'moment of disengagement' for churned customers.

2.  **Age Group Specific Churn Drivers:**
    *   **Observation:** Our analysis suggests varying churn rates across different age groups.
    *   **Investigation Required:** Delve deeper into the specific needs and pain points of the age groups exhibiting higher churn. Are their expectations not being met? Are our product offerings or marketing messages less resonant with them? Focus groups or targeted feedback sessions could provide valuable qualitative data.

3.  **Acquisition Channel Performance vs. Retention:**
    *   **Observation:** The acquisition channel appears to influence churn risk.
    *   **Investigation Required:** Analyze the customer lifetime value (CLTV) and 90-day retention rates specifically by acquisition channel. Some channels might be bringing in customers with lower intent or a poor fit for our services. This insight will guide future marketing spend optimization, ensuring we acquire customers who are more likely to stay.

4.  **Impact of Lower Total Spending:**
    *   **Observation:** Customers with lower total spending are more prone to churn.
    *   **Investigation Required:** Determine if lower spending is a cause or a symptom of disengagement. Can we identify specific product categories that fail to drive repeat purchases? Are these customers experiencing issues that deter further spending? Consider micro-segmentation of low-spending customers to understand their journey and potential barriers.

5.  **Root Cause Analysis for Critical Support Issues:**
    *   **Observation:** Certain types of support tickets (e.g., technical issues, billing problems) are associated with higher churn rates.
    *   **Investigation Required:** Conduct a thorough root cause analysis for these high-churn-risk issue types. Is there a systemic product flaw, a gap in our documentation, or an inconsistency in our support resolution process? Improving the resolution of these specific issues could have a significant impact on retention.

### Recommendation

By proactively investigating these areas, we can move beyond correlation to causation. This data-driven approach will enable us to design highly targeted, impactful retention campaigns that address the true underlying reasons for churn, rather than simply treating symptoms. This will maximize the return on investment for our retention efforts and foster a more loyal customer base.
